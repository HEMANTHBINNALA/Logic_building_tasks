```python
#Problem 1: Mobile Recharge Validation System

def validate_recharge(amount):
    valid_plans = [199, 299, 399, 599]

    if amount < 50:
        print("Recharge amount must be at least ₹50.")
        return False
    elif amount not in valid_plans:
        print("Invalid plan. Available plans are:", valid_plans)
        return False
    else:
        print(f"Recharge of ₹{amount} successful ")
        return True



while True:
    try:
        recharge_amount = int(input("Enter recharge amount: "))
        if validate_recharge(recharge_amount):
            break   # Exit loop on successful recharge
    except ValueError:
        print("Please enter a valid numeric amount.")
```

```python
#Problem 2: Inventory Reorder Alert System

def check_inventory(stock_data):
    for product, quantity in stock_data.items():
        if quantity < 15:
            print(f"{product}: Reorder Alert (Stock = {quantity})")
        else:
            print(f"{product}: Stock OK (Stock = {quantity})")



inventory = {
    "Rice": 15,
    "Sugar": 30,
    "Oil": 8,
    "Wheat": 40
}

check_inventory(inventory)
```

```python
#Problem 3: Student Result Processing System

def process_result(marks):
    total = 0
    for mark in marks:
        total += mark


    average = total / len(marks)

    if average >= 50:
        return f"Average: {average:.2f} - Pass"
    else:
        return f"Average: {average:.2f} - Fail"



student_marks = [60, 45, 70, 55]
result = process_result(student_marks)
print(result)
```

```python
#Problem 4: Cab Fare Estimator with Retry Option


def calculate_fare(distance_km, is_peak_hour):
    base_fare = 50
    per_km_rate = 12

    total_fare = base_fare + (distance_km * per_km_rate)

    if is_peak_hour:
        total_fare += total_fare * 0.25  # 25% extra

    return total_fare


# Retry logic
while True:
    try:
        distance = float(input("Enter distance in km: "))
        peak_input = input("Is it peak hour? (yes/no): ").lower()

        is_peak = peak_input == "yes"

        fare = calculate_fare(distance, is_peak)
        print(f"Total Cab Fare: ₹{fare:.2f}")

        retry = input("Do you want to calculate again? (yes/no): ").lower()
        if retry != "yes":
            break

    except ValueError:
        print("Invalid input. Please enter numeric distance.")
```

```python
#Problem 5: Employee Attendance Eligibility Checker

def check_attendance(attendance_list):
    total_days = len(attendance_list)
    present_days = 0

    # Count present days using loop
    for day in attendance_list:
        if day == "P":
            present_days += 1

    # Calculate attendance percentage
    attendance_percentage = (present_days / total_days) * 100

    # Check eligibility
    if attendance_percentage >= 75:
        return f"Attendance: {attendance_percentage:.2f}% - Eligible"
    else:
        return f"Attendance: {attendance_percentage:.2f}% - Not Eligible"


# Example usage
attendance = ["P", "P", "A", "P", "P", "A", "P", "P"]
result = check_attendance(attendance)
print(result)
```

```python
#Problem 6: Password Strength Checker

def check_password_strength(password):
    has_digit = False
    has_special = False
    special_chars = "@#$"

    # Rule 1: Check length
    if len(password) < 8:
        return "Weak Password  (Minimum 8 characters required)"

    # Rule 2 & 3: Loop through characters
    for char in password:
        if char.isdigit():
            has_digit = True
        if char in special_chars:
            has_special = True


    if has_digit and has_special:
        return "Strong Password "
    else:
        return "Weak Password (Must contain digit and special character)"

password = input("Enter password: ")
result = check_password_strength(password)
print(result)
```
